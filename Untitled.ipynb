```python
import keras as k
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.preprocessing.image import ImageDataGenerator
from keras.layers.normalization import BatchNormalization

import cv2
from tqdm import tqdm
import csv
import pandas as pd
import numpy as np
```

```python
x_train = []
x_test = []
y_train = []
y_test = []
```

```python
df_train = pd.read_csv(r'C:\Users\indva\Downloads\Workspace\CNN\circles.csv')
df_test = pd.read_csv(r'C:\Users\indva\Downloads\Workspace\CNN\testfile.csv')
df_train
```

```python
flatten = lambda l: [item for sublist in l for item in sublist]
labels = list(set(flatten([l.split(' ') for l in df_train['tags'].values])))
print(labels)
```

```python
label_map = {l: i for i, l in enumerate(labels)}
print(label_map)
```

```python
for f, tags in tqdm(df_train.values[:197], miniters=197):
    img = cv2.imread(r'C:\Users\indva\Downloads\Workspace\CNN\images\{}.jpg'.format(f))
    #print(img)
    img = cv2.resize(img, (28, 28))
    targets = np.zeros(2) 
    for t in tags.split(' '):
        targets[label_map[t]] = 1 
    x_train.append(img)
    y_train.append(targets)
```

```python
for f, tags in tqdm(df_test.values[:50], miniters=52):
        img = cv2.imread(r'C:\Users\indva\Downloads\Workspace\CNN\Test\{}.jpg'.format(f))
        #print(img)
        img = cv2.resize(img, (28, 28))
        filename = f  
        x_test.append(img)
        y_test.append(filename)  
          
```

```python
y_train = np.array(y_train, np.uint8)
#print(y_train)
x_train = np.array(x_train, np.float16) / 255.
x_test = np.array(x_test,np.uint8) / 255.
#print(x_test)
y_test = np.array(y_test,dtype='object')
```

```python
model = Sequential()
model.add(Conv2D(5, (3,3), input_shape=(28,28,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(5, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(1000, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(2, activation='softmax'))
#model.load_weights('Initial_weight')

model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])

              
model.fit(x_train, y_train, batch_size=20,epochs=30)

```

```python
model = Sequential()
model.add(Conv2D(28, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=(28, 28, 3),bias=False))
model.add(BatchNormalization(epsilon=1e-06, mode=0, momentum=0.9, weights=None))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, (3, 3), activation='relu',bias=False))
model.add(BatchNormalization(epsilon=1e-06, mode=0, momentum=0.9, weights=None))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(128, (3, 3), activation='relu',bias=False))
model.add(BatchNormalization(epsilon=1e-06, mode=0, momentum=0.9, weights=None))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(512, activation='relu',bias=False))
model.add(BatchNormalization(epsilon=1e-06, mode=0, momentum=0.9, weights=None))
model.add(Dropout(0.5))
model.add(Dense(2, activation='sigmoid'))

model.compile(loss='binary_crossentropy', # We NEED binary here, since categorical_crossentropy l1 norms the output before calculating loss.
              optimizer='adam',
              metrics=['accuracy'])
              
model.fit(x_train, y_train, batch_size=20,epochs=75)
```

```python
p_test = model.predict(x_test,batch_size=48)
print(p_test)

```

```python
step = 0
predict_list= []
while step < 1 :
    predictions= []
    for i in range(0,50):
        filename = y_test[i]
        probabilities = p_test[i]
        #print(sum(probabilities))
        mydict = {labels[0]: probabilities[0],labels[1]: probabilities[1]}
        tags=[]
        for classes,probabilities in mydict.items():
            if probabilities > 5e-01:
                tags.append(classes)
        final = {'img': filename,'tags': tags}           
        
        predictions.append(final)                  
    predict_list.extend(predictions)
    step += 1
  
print("Now Writing the File.")
toCSV = predict_list
keys = toCSV[0].keys()
with open(r'C:\Users\indva\Downloads\Workspace\CNN\predictions_.csv', 'a') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(toCSV)
print("Now written the file")
```
